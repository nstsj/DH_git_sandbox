# командная строка в Unix* (Linux, MacOS)

основные команды для Mac OS

* pwd - актуальная директория 
* ls - список файлов и папок в директории
* cd - смена директории (./ вперед   ../ назад)
* mkdir - создание новой директории
* touch - создание нового файла
* rm - удаление файла
* rm -r - удаление директории
* clear - очистка рабочего поля
* history - просмотр предыдущих команд

# работа с файловой системой через Python

В питоне, чтобы работать с файлами, операционной системой, перемещаться среди папок, существует модуль os. 
Подробнее о модуле можно почитать [тут](https://pythonworld.ru/moduli/modul-os.html).

Рассмотрим основные методы:

````python
# чтобы начать работать с модулем, его надо импортировать

import os
````

````python
#узнаем имя OS с которой работаем
os.name
````

````python
os.getcwd() # текущая рабочая директория (cwd = current working directory)
````

````python
os.listdir() # список файлов и директорий в папке
````

````python
os.mkdir("new") # создаёт новую директорию (прописываем путь в скобках)
````

````python
os.listdir() # еще раз посмотрим список файлов и директорий в этой папке
````

````python
os.chdir("./new") # сменяет текущую рабочую директорию, путь к новой нужной указываем в скобках
os.getcwd() #  проверим текущую директорию
````

````python
os.listdir("./new") # посмотрим, что в новой директории
````

````python
# удаление директории
os.chdir("../") # сначала выйдем из директории
os.rmdir('new') # удаляет пустую директорию.  
os.listdir() # посмотрим на список файлов снова,  new уже нет
````

````python
os.getcwd()
````

Пути сохраняются как строки,  наборы элементов в директории - как списки; с объектами можно совершать все стандартные питонские действия
